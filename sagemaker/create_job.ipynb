```python
import pandas as pd

loaded_df = pd.read_pickle('../data/social_explorer/5_dataset_expanded.pkl')

loaded_df.to_parquet('5_dataset_expanded.parquet')
```

```python
import sagemaker

sagemaker_session = sagemaker.Session()
sagemaker_session.upload_data('5_dataset_expanded.parquet')
```

```python
from sagemaker.pytorch import PyTorch

role = "arn:aws:iam::702497914099:role/service-role/AmazonSageMaker-ExecutionRole-20231130T083663"

input_data = 's3://sagemaker-us-east-1-702497914099/data/5_dataset_expanded.parquet'

estimator = PyTorch(
    entry_point='train.py',
    source_dir="./",
    role=role,
    framework_version='2.0',
    py_version='py310',
    instance_count=1,
    instance_type='ml.g5.2xlarge',
    hyperparameters={
        'batch-size': 16,
        'epochs': 1,
        'learning-rate': 2e-4,
        'model-name': 'NousResearch/Llama-2-7b-chat-hf',
    },
)

estimator.fit({'train': input_data})
```
